```python
import pickle
import lzma
import pandas as pd

# read the results from the pickle files

results_datetime = "2024_07_27_14_54_37"
model_type = "xgboost"
cross_val_type="custom_cross_val"

with lzma.open(f"reports/{model_type}/{cross_val_type}/performances/graphical_scores_{results_datetime}.pkl", "rb") as f:
    graph_perf = pickle.load(f)

num_perf = pd.read_csv(f"reports/{model_type}/{cross_val_type}/performances/numerical_scores_{results_datetime}.csv")

# get the keys of the graph_perf dictionary
temp_values = list(graph_perf.keys())
```

# Numerical performances

```python
num_perf
```

# Graphical performances

## training performances

```python
for temp in temp_values:
    graph_perf[temp]["training_scores"].update_layout(width=700, height=600).show()
```

## Validation scores

```python
for temp in temp_values:
    graph_perf[temp]["validation_scores"].update_layout(width=700, height=600).show()
```

## Test scores


```python
for temp in temp_values:
    graph_perf[temp]["testing_scores"].update_layout(width=700, height=600).show()
```
